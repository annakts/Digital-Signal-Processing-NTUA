```python
N=1024

def getIndexes(indexlist, N):
    freq = [0.5346, 0.5906, 0.6535, 0.7217, 0.9273, 1.0247, 1.1328]
    indexlist = [(x * 2 * np.pi / N) for x in indexlist]
    middle_points = [0]*8
    tones = set() 

    for i in range(6):
        x = (freq[i]+freq[i+1])/2
        middle_points[i+1] = x

    middle_points[0] = freq[0] - 0.028
    middle_points[7] = freq[6] + 0.055

    for k in indexlist:
        for i in range(7):
            if (k > middle_points[i] and k < middle_points[i+1]):
                k = freq[i]
                tones.add(k)
                break
    return tones

#testing
u = 1/(np.pi * 2)
ks = [0.53, 0.58, 0.65, 0.70, 0.92, 1, 1.1]
ks = [(x * u * N) for x in ks]

print(getIndexes(ks,N))
```

```python
def getTonesEnergy(signal, L=1000, N=1024):
    hamming = np.hamming(L)
    f = np.linspace(0, N-1, N)
    if(len(signal)>9000):
        length = 10
    else:
        length = 8
    start = 0
    signalsTones = np.zeros((L, length))
    for i in range(length):
        end = start + 1000    
        signalsTones[:, i] = hamming*(signal[start:end])
        start += 1100

    Energy_ST = np.zeros((N, length))

    for i in range(length):
        Energy_ST[:, i] = abs(np.fft.fft(signalsTones[:, i], N))**2
   
    return Energy_ST

def ttdecode(signal):
    energy_tones = getTonesEnergy(signal)
    Vector = []
    flag = False
    indexes = []
    length = energy_tones.shape[1]

    for i in range(length):
        indexes.append([])

    for j in range(length):
        goToSecond = False
        maxValue1 = 0
        maxValue2 = 0
        index1 = 0
        index2 = 0
        limit = 10000
        goToSecond = False
        for i in range(energy_tones.shape[0]):
            normalizedFreqs = []

            if goToSecond is False:
                if (energy_tones[i, j] > limit and not flag):
                    flag = True
                if flag:
                    if energy_tones[i, j] < limit:
                        goToSecond = True
                        flag = False
                    else:
                        if(maxValue1 < energy_tones[i, j]):
                            maxValue1 = energy_tones[i, j]
                            index1 = i
            elif goToSecond:
                if (energy_tones[i, j] > limit and not flag):
                    flag = True
                if flag:
                    if energy_tones[i, j] < limit:
                        flag = False
                        break
                    else:
                        if(maxValue2 < energy_tones[i, j]):
                            maxValue2 = energy_tones[i, j]
                            index2 = i 
        normalizedFreqs.append(index1)
        normalizedFreqs.append(index2)
        digits_freqs_set = getIndexes(normalizedFreqs,1024)
        digits_freqs = list(digits_freqs_set)

        index = 0
        w_row = np.array([0.5346, 0.5906, 0.6535, 0.7217])
        w_column = np.array([0.9273, 1.0247, 1.1328])
        w = [] 


        for i in range(10):
            w.append([])

        k = 1
        for i in range(3):
            for j in range(3):
                w[k].append(w_row[i])
                w[k].append(w_column[j])
                k += 1

        w[0].append(w_row[3])
        w[0].append(w_column[1])
        for freqsCouple in w:
            if digits_freqs == freqsCouple:
                Vector.append(index)
                break
            index += 1
    return Vector


print('Digits of AM:', ttdecode(signall))    
```
